```python
from bs4 import BeautifulSoup
# 파일 읽기
with open('greenData/member_data.xml','r',encoding='utf-8-sig') as f:
    # 파싱
    soup = BeautifulSoup(f,'xml')
# 데이터 추출 (select 모두 출력 : 내가 지정한 'member'(반드시 태그여야함) 안의 모든 것 

member = soup.select('member')

for m in member : 
    name = m.select_one('name').text
    age = m.select_one('age').text
    city = m.select_one('city').text
    print(f"이름:{name}, 나이:{age}, 도시:{city}")
```

```python
df = pd.DataFrame({
        "미국":[2.1,2.2,2.3],
        "한국":[0.4,0.5,0.45],
        "중국":[10,13,15]
    },index=[2000,2012,2013])
# 데이터프레임 데이터 접근하기
# df.loc : index를 통해서 값을 찾음
# df.iloc : 인덱스 번호를 통해서 값을 찾음(0부터 시작)
df.loc[2000] # index라는 데이터프레임의 레이블을 이용해 값을 출력
df.iloc[2] # 이미 가지고있는 0부터 시작하는 index로 값을 출력
print(df['한국'][2000]) # 특정한 key값의 데이터를 추출할때

#데이터프레임에 컬럼(=열=세로) 삭제 : del
#데이터 프레임에 행(=가로) 삭제 : df.drop([인덱스이름])
df.drop([2012])
```

```python
import pandas as pd

data ={
    '대여소':['강남','강남','홍대','홍대','잠실','잠실'],
    '이용자수':[120,150,80,90,200,180],
    '대여시간':[30,45,20,25,60,55],
    '이용일자':['2024-01-01','2024-01-02','2024-01-01','2024-01-02','2024-01-01','2024-01-02']
}

df = pd.DataFrame(data)

```

```python
#Data자료를이용해이용자수가100명이상인데이터만출력하시오

df[df['이용자수']>=100]
```

```python
#Data자료를이용해대여시간기준으로오름차순정렬하여출력하시오.
# 내림차순 df.sort_values(by='대여시간',ascending=False)
df.sort_values(by='대여시간')
```

```python
#Data자료를이용해"총이용량" 컬럼을이용자수* 대여시간으로계산하여추가하시오.
df['총이용량'] = df['이용자수']*df['대여시간']
df
```

```python
#문제 : Data자료를이용해컬대여소별총이용자수구하여출력하시오.

df.groupby('대여소')['이용자수'].sum().sort_values(ascending=False)
```

```python
#Data자료를이용해이용자수가100 이상인데이터만추출하여 대여소별 평균 대여시간 구하여 출력하시오.
result=df[df['이용자수']>=100]
result.groupby('대여소')['대여시간'].mean()
```

```python
#문제 : Data 자료를 이용해 대여소별 "총이용량" 합계 구하기
df.groupby('대여소')['총이용량'].sum()
```

```python
#문제 : 가장 바쁜 날짜 찾기
result=df.groupby('이용일자')['이용자수'].sum().sort_values(ascending=False)
print(result.index[0])
```

```python
#문제 : 이용자수 평균이 가장 높은 대여소 찾기.
result=df.groupby('대여소')['이용자수'].mean().sort_values(ascending=False)
print(result)
print(result.index[0])
```
