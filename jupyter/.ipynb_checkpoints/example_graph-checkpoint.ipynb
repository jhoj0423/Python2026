• 과목별 점수를 세로 막대그래프로 표현하시오 
• 그래프 제목: "과목별 점수" 
• x축: 과목 / y축: 점수 
• 막대 위에 점수 값 표시
• 아래 데이터를 이용해 작성하시오.
df = pd.DataFrame({
    "과목": ["Java", "Python", "React", "SQL"],
    "점수": [85, 90, 75, 80]
})

```python
#반드시 pandas는 import해서 사용

import pandas as pd
# matplotlib 매트플롯립도 반드시 import해서 사용한다.
import matplotlib.pyplot as plt

#반드시 한글 깨짐 현상 코드를 삽입한다.

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(6,5))




df = pd.DataFrame({
    "과목": ["Java", "Python", "React", "SQL"],
    "점수":[85,90,75,80]
})

bars = plt.bar(df['과목'],df['점수'],width = 0.6,color='skyblue')

## 변경 문제 ) Python 항목 부분에만 레이블 값이 출력 되도록 수정하시오
for i in range(len(bars)):
    bar = bars[i]
    if df['과목'][i] == "Python":
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, height, f'{int(height)}',ha = 'center',va='bottom')


plt.title('과목별 점수')
plt.xlabel('과목')
plt.ylabel('점수')

plt.show()
```

```python
#반드시 pandas는 import해서 사용

import pandas as pd
# matplotlib 매트플롯립도 반드시 import해서 사용한다.
import matplotlib.pyplot as plt


df = pd.DataFrame({
    "이름": ["김학생", "이학생", "박학생", "최학생"],
    "공부시간": [3, 5, 2, 4]
})
#반드시 한글 깨짐 현상 코드를 삽입한다.

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(6,3))

df2=df[df['이름'] != '김학생']


plt.barh(df2['이름'],df2['공부시간'],height=0.4,color="skyblue")

plt.title('학생별 공부시간')
plt.xlabel('공부시간')
plt.ylabel('이름')

plt.show()
```

### 6번 문제
#### 두 데이터를 부서 기준으로 병합하시오(outer 조인)
#### 곁측치는 각 컬럼의 평균값으로 처리하기오
#### 기본급과 성과급 수당을 이용하여 누적 세로 맥대 그래프를 생성하시오
#### 그래프 구성 :
#### 아래: 기본급
#### 위 : 성과급 (누적) 
#### 범례(legend ) 추가
#### 그래프 제목: "부서별 급여 구조"
#### x축: 부서 / y축: 금액
#### 그래프를 이미지로 저장 (salary_chart.png, dpi=300)

```python
df1 = pd.DataFrame({
    "부서": ["개발팀", "영업팀", "인사팀"],
    "기본급": [300, 250, 200]
})
df2 = pd.DataFrame({
    "부서": ["개발팀", "영업팀", "마케팅팀"],
    "성과급": [100, 80, 70]
})
df3 = pd.DataFrame({
    "부서": ["개발팀", "영업팀", "인사팀"],
    "수당": [50, 40, 30]
})
merged = pd.merge(df1, df2, on="부서",how="outer")
merged2 = pd.merge(merged, df3, on="부서",how="outer")
merged2['기본급'] = merged2['기본급'].fillna(merged2['기본급'].mean())
merged2['성과급'] = merged2['성과급'].fillna(merged2['성과급'].mean())
merged2['수당'] = merged2['수당'].fillna(merged2['수당'].mean())

merged2
```
