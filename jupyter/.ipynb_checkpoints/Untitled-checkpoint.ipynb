```python
import pandas as pd
import pymysql

# DB 연결 설정
conn = pymysql.connect(host='localhost', user='root', password='12345678', db='resort', charset='utf8mb4')

# 쿼리 실행 및 데이터프레임 저장
df = pd.read_sql_query("SELECT * FROM hotel", conn)

# CSV로 저장 (인코딩을 utf-8-sig로 하면 엑셀에서도 바로 잘 보입니다)
df.to_csv("greenData/my_data.csv", index=False, encoding="utf-8-sig")

conn.close()
```
