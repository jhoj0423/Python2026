```python
import pandas as pd
# matplotlib 매트플롯립도 반드시 import해서 사용한다.
import matplotlib.pyplot as plt
# 숫자 계산 쉽게하는 라이브러리 불러오기
import numpy as np
#데이터 사이를 부드럽게 열결해주는 함수 가져오기
from scipy.interpolate import make_interp_spline

#df1 = pd.read_csv('greenData/resort/hotel.csv',encoding='utf-8-sig') # 호텔의 데이터
df2 = pd.read_csv('greenData/resort/reservation.csv',encoding='utf-8-sig') # 예약의 데이터
df3 = pd.read_csv('greenData/resort/room.csv',encoding='cp949')

# 가장 인기 있는 호텔 가격대별
merged = pd.merge(df3,df2,on="r_code",how='inner')
list01= merged.groupby('h_code').size()
list02= merged.groupby('h_code')['price'].min().sort_values(ascending=False)

li01 = pd.DataFrame(list01)
li02 = pd.DataFrame(list02)


merged02 = pd.merge(li01,li02,on='h_code',how='inner')
num1 = merged02[merged02['price']<=100000].sum()
num2 = merged02[(merged02['price']>=100000) & (merged02['price']<=200000)].sum()
num3 = merged02[(merged02['price']>=200000) & (merged02['price']<=300000)].sum()

li = ({
    "가격대":['100000원이하','200000원이하','300000원이하']
})
li03 = pd.DataFrame(li)
li03['예약자수'] = num1[0]
li03['예약자수'] = num2[0]
li03['예약자수'] = num3[0]
li03
```
