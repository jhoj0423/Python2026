## utf-8-sig는 윈도우 환경에서 텍스트 파일 저장시 앞 부분에 기호가 같이 저장되는 부분을 해결해주는 인코딩이다.

```python
with open('greenData/text_data.txt','w',encoding='utf-8-sig') as f:
    f.write('데이터 분석을 위한 첫 걸음\n')
    f.write('파이썬 기본 파일 입출력 실습입니다.\n')
    f.write('1.기초 문법 익히기\n')
    f.write('2라이브러리 활용하기\n')
    f.write('3.시각화 진행하기\n')
```

## DictReader => csv 파일을 읽기

```python
import csv
# 1.파일 읽기
with open('greenData/score_data.csv','r',encoding='utf-8-sig') as f:
    # 2.
    score = csv.DictReader(f)
    print(score)
    # 3. 헌쥴 싹 확인 하면서 if 조건 검사
    for i in score:
        # 파일에서 읽어온 도튼 숫자는 문자열이름로 반드시 int()로 숫자형으로 자료를 변환한다
        if int(i['국어'])>=85:
            print(i)
            
    
```
